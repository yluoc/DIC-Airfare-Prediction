Above again, it underlines the need for preprocessing in that such processing not only increases the quality of data in a lot of cases but also generally results in more accurate and reliable models. Preprocessing a dataset means allowing us to remove the noise from our models or elements which may not very well be relevant, and in essence, they allow better generalization to new, unseen data. With proper preprocessing, we are also in a position to reduce the computation complexity involved in dealing with our models and hence improve efficiency.

In overall consideration, it involves preprocessing data—a very vital stage during the preparation of information for analysis or application in machine learning. It allows models to be created, which produce results that are more precise and reliable.

## Data Cleaning Step 6

In this step, we convert the duration of each flight from the traditional hour-minute format into a unified minute format. This transformation facilitates uniformity in the representation of flight durations, simplifying subsequent calculations and analyses. Converting the duration to minutes allows for easier comparison and computation across different flights, regardless of their original time units. This standardized format eliminates the need for complex conversions during analysis and streamlines the process.

```python
# Convert time_taken format from  hour-minute to minute
def convert_to_minutes(time_str):
    parts = time_str.replace('h', '.').replace('m', '.').split('.')
    hour = int(parts[0])
    min = 0
    if len(parts):
        min = int(parts[1])
    return hour*60 + min

df['time_taken'] = df['time_taken'].apply(convert_to_minutes)

df
```

## Data Cleaning Step 7

We identified and removed outliers from the dataset after conducting a comparative analysis between airfare and flight duration. Outliers are data points that deviate significantly from the majority of observations and can skew the results of our analysis. Here, outliers represent extreme values that distort the relationship between the variables. The removal of outliers helps improve the robustness and accuracy of our analysis, allowing us to identify meaningful patterns and trends without the influence of extreme values.

- Filter and Remove these outliers by using Inter-quartile range formula.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data = df['price'])

plt.show()

sns.boxplot(data = df['time_taken'])

plt.show()
```

```python
def get_outliers(df, dis):
    # Define inter-quartile range
    q1 = df[dis].quantile(0.25)
    q3 = df[dis].quantile(0.75)
    iqr = q3 - q1

    lower_boundary = q1 - iqr*1.5
    upper_boundary = q3 + iqr*1.5

    # Define outliers
    Outliers = df.index[(df[dis] < lower_boundary) | (df[dis] > upper_boundary)]

    return Outliers

outlier_list1, outlier_list2 = [], []
outlier_list1.extend(get_outliers(df, 'price'))
outlier_list2.extend(get_outliers(df, 'time_taken'))

def remove_outliers(df, ls):
    ls = sorted(set(ls))
    df = df.drop(ls)
    return df
# Remove outliers in price
df = remove_outliers(df, outlier_list1)
# Remove outliers in time_taken
df = remove_outliers(df, outlier_list2)

df.shape
```

## Data Cleaning Step 8

We transformed the ticket-booking date for each flight into a new variable called "days-left." We selected a reference date, "10-02-2022," and designated it as day 0. Using this reference point, we calculated the number of days left until the flight departure date for each booking.

By converting the ticket-booking date into days-left format, we effectively capture the temporal proximity of each booking to the departure date. This transformation facilitates temporal analysis and allows us to explore patterns related to booking behavior in relation to the flight departure date.

This conversion of ticket-booking dates into days-left format serves as a valuable feature in predicting airfare. By incorporating the temporal proximity of booking to the flight departure date, our predictive models can capture potential patterns or trends in airfare fluctuations over time. For instance, we hypothesize that airfare may vary based on the remaining days until departure, with prices potentially increasing closer to the departure date due to higher demand or limited seat availability.

```python
from datetime import datetime

def calculate_days_diff(date):
    # Define the reference date
    reference_date = datetime.strptime("10-02-2022", "%d-%m-%Y")
    # Convert the input date to datetime format if it's not already
    if not isinstance(date, datetime):
        date = datetime.strptime(date, "%d-%m-%Y")
    # Calculate the difference in days
    days_diff = (date - reference_date).days
    return days_diff


# Convert the Date column to datetime format and apply the function
df['Days_Left'] = df['date'].apply(calculate_days_diff)

df_dropped_column = df.drop(columns=['date'],axis=1, inplace=True)

df
```

## Data Cleaning Step 9

We are binning the departure and arrival times of the flights. Binning involves categorizing continuous variables, such as time, into discrete intervals or "bins." This allows us to simplify the data and identify patterns or trends more effectively. By binning departure and arrival times, we group flights into time intervals based on their scheduled departure and arrival times. This categorization enables us to analyze the data in a more structured and interpretable manner, as we can observe trends across specific time periods rather than individual timestamps.

Binning also helps to reduce the impact of small variations in departure and arrival times, making it easier to identify broader patterns in flight scheduling and timing. Additionally, binning allows for easier visualization and comparison of flights across different time intervals.

Following are the bins that we have chosen:
- Early Morning
- Morning
- Afternoon
- Evening
- Night

```python
def categorize_time_of_day(time_str):
# Convert the time to a datetime object
    time = pd.to_datetime(time_str, format='%H:%M').time()
    
    # Define the time bins
    bins = [(0, 4), (4, 8), (8, 12), (12, 16), (16, 20), (20, 24)]
    labels = ['Night', 'Early Morning', 'Morning', 'Afternoon', 'Evening', 'Night']
    
    # Calculate the hour of the time
    hour = time.hour + (time.minute / 60)

    # Determine the time category based on the hour
    for (start, end), label in zip(bins, labels):
        if start <= hour < end:
            return label
    # Handle the edge case for times between 00:00 and 04:00
    return 'Night'
```

```python
df['dep_time_category'] = df['dep_time'].apply(categorize_time_of_day)
df['arr_time_category'] = df['arr_time'].apply(categorize_time_of_day)
```
